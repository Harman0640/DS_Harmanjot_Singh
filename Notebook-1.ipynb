 ### EDA

#### Fear / Greed Index Value Distribution
Histogram / KDE

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.histplot(df2['value'] , kde=True , palette='steelblue')
plt.title("Fear / Greed Index Value Distribution")
```

#### Sentiment Classification Distribution
Bar Plot / Count Plot

```python
sns.countplot(x='classification' , data=df2 , palette='pastel')
plt.title("Sentiment Classification Distribution")
```

#### Trade Side Distribution (BUY / SELL)
Bar Plot (Count Plot)

```python
sns.countplot(x='Side' , data=df1 , palette='coolwarm')
```

### Sentiment Value vs Closed PnL

```python
sns.scatterplot(x='value', y='Closed PnL', data=new_df)
```

### Total Closed PnL by Sentiment

```python
sns.barplot(x='classification', y='Closed PnL', data=new_df , palette='pastel')
plt.title("Total Closed PnL By Sentiment")
```

#### Average Trade Size by Sentiment

```python
sns.barplot(x='classification', y='Size USD', data=new_df)
plt.title("Average Trade Size By Sentiment")
```

### Calciulate Daily Avg PnL By Sentiment

```python
daily_avg_pnl = (new_df.groupby(['Trade_date', 'classification'])['Closed PnL'].mean().reset_index())
daily_avg_pnl
```

```python
plt.figure(figsize=(12, 6))

sns.lineplot(data=daily_avg_pnl , x='Trade_date' , y='Closed PnL', hue='classification')

plt.title('Daily Average Closed PnL by Sentiment Classification')
plt.xlabel('Trade Date')
plt.ylabel('Average Closed PnL')
# plt.xticks(rotation=0)

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
